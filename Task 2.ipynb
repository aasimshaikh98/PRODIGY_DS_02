```python
#unique values in the column "No of Times Visited"
Raw_Housing_Data['No of Times Visited'].unique()
```

```python
#replacing the nan values with zero
Raw_Housing_Data['No of Times Visited'] = Raw_Housing_Data['No of Times Visited'].fillna('Zero')
```

```python
#also mapping it to convert the categorical variable to a numerical variable to gain more insights
mapping = {'Zero':0,
          'Once':1,
          'Twice':2,
          'Thrice':3,
          'Four':4}
Raw_Housing_Data['No of Times Visited']=Raw_Housing_Data['No of Times Visited'].map(mapping)
```

```python
#unique values after conversion
Raw_Housing_Data['No of Times Visited'].unique()
```

```python
#displaying data
Raw_Housing_Data.head()
```

```python
#Sale price in the dataset in a visualized way
plt.plot(Raw_Housing_Data['Sale Price'],color='red')
plt.xlabel("Record Number")
plt.ylabel("Sale Price")
plt.show()
```

```python
#displaying the count of each condition in the house
Raw_Housing_Data.groupby('Condition of the House')['ID'].count()
```

```python
#assigning values and label for visualization
values=(30,1701,14031,5679,172)
labels=('Bad','Excellent','Fair','Good','Okay')
```

```python
#pie chart to visualize condition of the house
plt.pie(values,labels=labels)
```

```python
#bar chart to visualize condition of the house
plt.bar(labels,values,color='Purple',linewidth=5,linestyle='dashed')
plt.xlabel("Condition of the House")
plt.ylabel("Count of the House")
plt.title("My First Graph")
plt.show()
```

```python
#scatter plot to identify relationship between "Flat Area (in sqft)" and "Sale Price"
#the insights tells us that as the flat area increases, the selling price also increases
plt.scatter(x = Raw_Housing_Data['Flat Area (in Sqft)'], y = Raw_Housing_Data['Sale Price'], color = 'blue')
plt.xlabel("Area")
plt.ylabel("Selling Price")
plt.title("Selling Price vs Area")
plt.show()
```

```python
#displaying the mean sale price in each condition of the house:
#initialising a new coloumn
Raw_Housing_Data['condition_sale']=0
#calculating mean based on condition of the house
for i in Raw_Housing_Data['Condition of the House'].unique():
    Raw_Housing_Data['condition_sale'][Raw_Housing_Data['Condition of the House']==str(i)]=Raw_Housing_Data['Sale Price'][Raw_Housing_Data['Condition of the House']==str(i)].mean()
#plotting the mean sales based on condition of the house
plt.figure(dpi=100)
plt.bar(Raw_Housing_Data['Condition of the House'].unique(),Raw_Housing_Data['condition_sale'].unique())
plt.xlabel("Condition of the House")
plt.ylabel("Mean Sale Price")
plt.show()
```
